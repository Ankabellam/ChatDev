## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyse and visualise data. The code cell below uses <strong>numpy</strong> to generate some random data, and uses <strong>matplotlib</strong> to visualise it. To edit the code, just click the cell and start editing.

```
# read csv file from local drive
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
# Read the CSV file from local drive
df = pd.read_csv('/content/bank_sample/bank.csv')
print(df.head())
```

```
#identify the features data types before entering into analysis

df.info()
print(df.dtypes)
```

```
#convert the datatypes using user-defined functions to concert the columns to appropriate types

def convert_to_float(value):
  try:
    return float(value)
  except ValueError:
    return value

def convert_to_int(value):
  try:
    return int(value)
  except ValueError:
    return value

def convert_to_category(value):
  return str(value)

for column in df:
  if df[column].dtype == 'object':
    df[column] = df[column].apply(convert_to_category)
  elif df[column].dtype == 'float64':
    df[column] = df[column].apply(convert_to_float)
  else:
    df[column] = df[column].apply(convert_to_int)

df.info()
```

```
#find and remove missing values

df = df.dropna()
df.head()
missing_values =df.isnull().sum()
print("The Missing values are",missing_values)
```

```
# prompt: visuvalize the missing values

#import missingno as msno
#msno.matrix(df)

import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.isnull(), cbar =False,cmap='viridis')
plt.show()
#drop missing values optional -> already dropped in above code cell
df =df.dropna()
```

```
#Average Balance of subscribed vs non-subscribed customers

subscribed_customers = df[df['y'] == 'yes']
non_subscribed_customers = df[df['y'] == 'no']

average_balance_subscribed = subscribed_customers['balance'].mean()
average_balance_non_subscribed = non_subscribed_customers['balance'].mean()

print("Average balance of subscribed customers:", average_balance_subscribed)
print("Average balance of non-subscribed customers:", average_balance_non_subscribed)
```

```
#Visualize average balance subscribed vs non-subscribed customers

import matplotlib.pyplot as plt
import seaborn as sns

# Create a bar chart
sns.barplot(x=['Subscribed', 'Non-Subscribed'], y=[average_balance_subscribed, average_balance_non_subscribed])

# Add title and labels
plt.title('Average Balance of Subscribed vs Non-Subscribed Customers')
plt.xlabel('Subscription Status')
plt.ylabel('Average Balance')

# Show the plot
plt.show()
```

```
#pie plot for the distribution of education

import matplotlib.pyplot as plt
education_distribution = df['education'].value_counts()

# Create a pie chart
plt.pie(education_distribution, labels=education_distribution.index, autopct='%1.1f%%')

# Add title
plt.title('Distribution of Education Levels')

# Show the plot
plt.show()
```

```
# create a ' season' feature from the 'month' column

df['season'] = df['month'].apply(lambda x: 'spring' if x in ['mar', 'apr', 'may'] else 'summer' if x in ['jun', 'jul', 'aug'] else 'fall' if x in ['sep', 'oct', 'nov'] else 'winter')

print(df[['month','season']].head())
```

```
#count plot for 'season' and 'y'

import matplotlib.pyplot as plt
sns.countplot(x = 'season', hue = 'y', data = df)
plt.title('Subscription Status by Season')
plt.show()
```

```
#Effect of 'Pdays' on subscription using bar plot

import matplotlib.pyplot as plt
# Group data by Pdays and subscription status
grouped_data = df.groupby('pdays')['y'].value_counts(normalize=True).unstack()

# Create a bar plot
grouped_data.plot(kind='bar', stacked=True)

# Add title and labels
plt.title('Effect of Pdays on Subscription')
plt.xlabel('Pdays')
plt.ylabel('Number of Customers')

# Show the plot
plt.show()
```

```
#Replace -1 in 'Pdays' with NaN and repeat analysis

import numpy as np
import pandas as pd
from google.colab import drive
import seaborn as sns
import matplotlib.pyplot as plt

# read csv file from local drive
drive.mount('/content/drive')
# Read the CSV file from local drive
df = pd.read_csv('/content/bank_sample/bank.csv')

# Replace -1 in 'Pdays' with NaN
df['pdays'] = df['pdays'].replace(-1, pd.NA)

# Find and remove missing values
df = df.dropna()

# Effect of 'Pdays' on subscription using bar plot

# Group data by Pdays and subscription status
grouped_data = df.groupby('pdays')['y'].value_counts(normalize=True).unstack()

# Create a bar plot
grouped_data.plot(kind='bar', stacked=True)

# Add title and labels
plt.title('Effect of Pdays on Subscription with NaN')
plt.xlabel('Pdays')
plt.ylabel('Number of Customers')

# Show the plot
plt.show()
```

```
# prompt: Analyze if customers with less than 0  balance take the subscription
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from google.colab import drive
# Create a subset of customers with balance less than 0
df_negative_balance = df[df['balance'] < 0]

# Calculate the subscription rate for customers with negative balance
subscription_rate_negative_balance = df_negative_balance['y'].value_counts(normalize=True)['yes']

# Create a subset of customers with balance greater than or equal to 0
df_non_negative_balance = df[df['balance'] >= 0]

# Calculate the subscription rate for customers with non-negative balance
subscription_rate_non_negative_balance = df_non_negative_balance['y'].value_counts(normalize=True)['yes']

# Print the subscription rates
print("Subscription rate for customers with negative balance:", subscription_rate_negative_balance)
print("Subscription rate for customers with non-negative balance:", subscription_rate_non_negative_balance)

# Compare the subscription rates
if subscription_rate_negative_balance > subscription_rate_non_negative_balance:
  print("Customers with negative balance are more likely to subscribe.")
else:
  print("Customers with non-negative balance are more likely to subscribe.")

#Bar Plot
subscription_rate_negative_balance.plot(kind='bar')
plt.title('Subscription proportion for  Customers with Negative Balance')
plt.xlabel('Subscription Status')
plt.ylable('No of customers(proportion)')
plt.show()
```
